## 1️⃣ Creating Lists

Lists can hold any data type and are used to store multiple items in a single variable.

```python
# Creating lists in different ways
# Lists can contain Different data types
groceries = ["milk", "bread", "eggs", "butter"]
groceries = ["potato", "bread", "bread", "eggs", "butter"] # List with duplicate values
scores = [95, 88, 76, 90] 
student = ["Rahul", 21, "B.Tech", 8.4]
empty_list = []
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
numbers = list((10, 20, 30, 40))
print(groceries, scores, student, empty_list, matrix, numbers)
```

```python
# Checking type of list
print(type(groceries))
```

```python
# List constructor to make list
mylist = list(("apple", "banana", "cherry"))	
print(mylist) 	
```

## 2️⃣ Accessing Elements (Indexing)

Access list elements using positive or negative indexes.

```python
movies = ["Inception", "Avatar", "Interstellar", "Tenet"]
print(movies[0])

print(movies[2])
print(movies[-1])
print(movies[-2])
```

## 3️⃣ Slicing Lists

Extract sub-lists using slicing `[start:end:step]`.

```python
# Will give a new list
tasks = ["wake up", "brush", "exercise", "study", "breakfast", "work"]
print(tasks[1:4])
print(tasks[:3])
print(tasks[3:])
print(tasks[::2])
print(tasks[::-1])
```

## 4️⃣ Updating List Elements

Change one or multiple elements of a list.

```python
prices = [100, 200, 300, 400]
prices[1] = 250
prices[1:3] = [220, 320]
prices[1:2] = [210, 215, 220]
print(prices)
```

## 5️⃣ Adding Elements

Add elements using append(), insert(), and extend().

```python
# append()
fruits = ["apple", "banana"]
fruits.append("mango")
fruits.append(["kiwi", "melon"])
print(fruits)

# insert()
subjects = ["Math", "Science", "History"]
subjects.insert(1, "English")
subjects.insert(0, "Computer")
print(subjects)

# extend()
menu = ["idli", "dosa"]
menu.extend(["poha", "upma"])
menu.extend(("tea", "coffee"))
print(menu)
```

## 6️⃣ Removing Elements

Remove elements using remove(), pop(), del, and clear().

```python
cars = ["BMW", "Audi", "Tesla", "Audi"]
cars.remove("Audi")
print(cars)

queue = ["A", "B", "C", "D"]
queue.pop()
queue.pop(0)
print(queue)

students = ["Ravi", "Neha", "Kiran", "Asha"]
del students[2]
del students[1:3]
print(students)

cart = ["item1", "item2", "item3"]
cart.clear()
print(cart)
```
